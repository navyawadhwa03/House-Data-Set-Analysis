```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
house=pd.read_csv('train.csv')
```

#### Hypothesis that we are going to cover in this analysis is as follows-

1. Determine if the sale price of a house is affected by style of house.
2. Determine if the year of remodelling has any effect on the sale price of the house.
3. Determine if the sale price of the house varies with Land Contour.
4. Determine if the sale price of the house is affected by number of full bathrooms.
5. Determine if the garage area changes with the year the garage was built.
6. Determine the trend of house sale over the years.

#### 1. Determine if the sale price of a house is affected by style of house.

```python
sns.barplot(x=house['HouseStyle'],y=house['SalePrice'])
plt.xlabel("Style of house")
plt.ylabel("Sale price")
plt.title("Sale price of the house based on the style of house")
```

#### 2. Determine if the year of remodelling has any affect on the sale price of the house.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearRemodAdd', y='SalePrice', data=house)
plt.title('Sale Price vs. Remodel date')
plt.xlabel('Date of Remodel')
plt.ylabel('Sale Price')
plt.show()
```

#### 3. Determine if the sale price of the house varies with Land Contour.

```python
sns.barplot(x=house['LandContour'],y=house['SalePrice'])
plt.xlabel("Land Contour")
plt.ylabel("Sale Price")
plt.title("Sale price vs Land Contour")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=house, x='LandContour', y='SalePrice')
plt.title('Sale Price vs Land Contour')
plt.xlabel('Land Contour')
plt.ylabel('Sale Price')
plt.xticks(rotation=45)
plt.show()
```

#### 4. Determine if sale price of the house is affected by number of full bathrooms

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=house, x='FullBath', y='SalePrice')
plt.title('Sale Price vs Full Bathroom')
plt.xlabel('Full Bathroom')
plt.ylabel('Sale Price')
plt.show()
```

#### 5. Determine if the garage area changes with the year the garage was built.

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(x='GarageYrBlt', y='GarageArea', data=house)
plt.title('Garage Area vs Year the garage was built')
plt.xlabel('Garage Year Built')
plt.ylabel('Garage Area')
plt.show()
```

#### 6. Determine the trend of house sales over the years.

```python
count_sales=house.groupby('YrSold')['YrSold'].count()
year_list = count_sales.index.tolist()
sales_counts = count_sales.tolist()
plt.figure(figsize=(8, 3))
plt.plot(year_list, sales_counts, label='Number of sales over the years')
plt.xlabel('Year Sold')
plt.ylabel('Number of Sales')
plt.title('Number of Sales Over the Years')
plt.legend()
plt.show()
```

```python
count_sales = house.groupby('YrSold')['YrSold'].count()
max_year = count_sales.idxmax()
min_year = count_sales.idxmin()
print("Maximum sales were in:",max_year)
print("Minimum sales were in:",min_year)
```
